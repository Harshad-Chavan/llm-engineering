```python
from openai import OpenAI
import os
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
ollama_url = "http://localhost:11434/v1"
gpt = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
ollama = OpenAI(api_key="ollama", base_url=ollama_url)
```

```python
gpt_sys_prompt = """"""
ollama_sys_prompt = """"""

gpt_messages = ['hi']
ollama_messages = ['hi there']

def call_gpt():
    # Setting the system prompt for gpt
    messages = [{"role": "system", "content": gpt_sys_prompt}]
    for gptm, olam in zip(gpt_messages, ollama_messages):
        messages.append({"role": "assistant", "content": gptm})
        messages.append({"role": "user", "content": olam})
    response = gpt.chat.completions.create(model='gpt-5-mini', messages=messages)
    return response.choices[0].message.content


def call_olama():
    messages = [{"role": "system", "content": ollama_sys_prompt}]
    for gptm, olam in zip(gpt_messages, ollama_messages):
        messages.append({"role": "user", "content": gptm})
        messages.append({"role": "assistant", "content": olam})
    messages.append({"role": "user", "content": gpt_messages[-1]})
    response = ollama.chat.completions.create(model="llama3.2:1b", messages=messages)
    return response.choices[0].message.content

for i in range(2):
    gpt_next = call_gpt()
    print(f"Poor farmer : {gpt_next}",end="\n")
    gpt_messages.append(gpt_next)
    olama_next = call_olama()
    print(f"Rich Farmer:{olama_next}",end="\n")
    ollama_messages.append(olama_next)
```
